```python
#importing libraries
import dlib
import numpy as np
from scipy import misc
import cv2 

# Loding models
det_face = dlib.get_frontal_face_detector()  #returns the default face detector of dlib

#dlib.shape_predictor is a tool that takes in an image region containing some object and outputs a set of point locations that define the pose of the object. 
pose_est = dlib.shape_predictor('/content/gdrive/My Drive/Im beside you./shape_predictor_68_face_landmarks.dat') #loading the pre trained face detector model
face2script = dlib.face_recognition_model_v1('/content/gdrive/My Drive/Im beside you./dlib_face_recognition_resnet_model_v1.dat') #This model is a ResNet network with 29 conv layers

#defining function to detect face in an image
def face_find(image, n=1):
 return det_face(image, n)

#defining function to encode a face which was detected by above face_find function 
def encoding(f_image,n=1):
 pointers_face = face_find(f_image)
 pose = pose_est
 prd = [pose(f_image, i) for i in pointers_face] 
 return [np.array(face2script.compute_face_descriptor(f_image, j, n)) for j in prd][0]  
```

```python
import os
import pandas as pd

# specifying paths to image directories. There are three folders given, each containing images of zoom meetings
path1='/content/gdrive/MyDrive/Im beside you./face_images/a055a66f-46ea-4c57-8de4-f3cc4f27c59c'
path2='/content/gdrive/MyDrive/Im beside you./face_images/dd40a338-faf9-481d-996c-6fe0f648d0b7'
path3='/content/gdrive/MyDrive/Im beside you./face_images/f5845e49-256f-4d70-9214-47bd5358c9d1'
dir1=os.listdir(path1)
dir2=os.listdir(path2)
dir3=os.listdir(path3)
# df=pd.DataFrame(index=np.arange(0), columns=np.arange(128))
```

```python
#Counting number of images
print('Images in 1st directory = ',len(dir1)) 
print('Images in 2nd directory = ',len(dir2)) 
print('Images in 3rd directory = ',len(dir3)) 
print('Toatl raw images = ',len(dir1)+len(dir2)+len(dir3)) 
```

```python
#Finding the images with no faces or having mask on it. and creating a log of all these images. 
no_face_list = [] #list where names of all no-face images be kept

#Defining empty data frames to store data-vector of each image from all three directories
df1=pd.DataFrame(index=np.arange(0), columns=np.arange(129))
df2=pd.DataFrame(index=np.arange(0), columns=np.arange(129))
df3=pd.DataFrame(index=np.arange(0), columns=np.arange(129))

#Accessing each directory one by one to find no-face images

#For 1st directory
no_face1=0 #initializing counting from zero
j=0
print('Going into 1st directory')
for i in dir1:
  image_path = path1+'/'+i
  known_image = cv2.imread(image_path) #reading image
  face=face_find(known_image) #finding face
  if len(face)==0: #if no face gets detected
    no_face1+=1
    no_face_list.append(i)

  else:                          #condition when model has successfully detected face
     enc = encoding(known_image) #extracting 128-D feature vector from an image 
     enc=enc.tolist() 
     enc.append(i)               #adding name of the image to the list for ease
     df1_=pd.DataFrame(enc) 
     df1_=df1_.transpose()
     df1=df1.append(df1_)        #appending features of a face to global dataframe
  j=j+1
  if j%1000 == 0:
    print(j,'...')  



#For 2nd directory
no_face2=0 #initializing counting from zero
j=0
print('Going into 2nd directory')
for i in dir2:
  image_path = path2+'/'+i
  known_image = cv2.imread(image_path) #reading image
  face=face_find(known_image) #finding face
  if len(face)==0: #if no face gets detected
    no_face2+=1
    no_face_list.append(i)

  else:                          #condition when model has successfully detected face
     enc = encoding(known_image) #extracting 128-D feature vector from an image 
     enc=enc.tolist() 
     enc.append(i)               #adding name of the image to the list for ease
     df2_=pd.DataFrame(enc) 
     df2_=df2_.transpose()
     df2=df2.append(df2_)        #appending features of a face to global dataframe
  j=j+1
  if j%1000 == 0:
    print(j,'...') 


#For 3rd directory
no_face3=0 #initializing counting from zero
j=0
print('Going into 3rd directory')
for i in dir3:
  image_path = path3+'/'+i
  known_image = cv2.imread(image_path) #reading image
  face=face_find(known_image) #finding face
  if len(face)==0: #if no face gets detected
    no_face3+=1
    no_face_list.append(i)

  else:                          #condition when model has successfully detected face
     enc = encoding(known_image) #extracting 128-D feature vector from an image 
     enc=enc.tolist() 
     enc.append(i)               #adding name of the image to the list for ease
     df3_=pd.DataFrame(enc) 
     df3_=df3_.transpose()
     df3=df3.append(df3_)        #appending features of a face to global dataframe
  j=j+1
  if j%1000 == 0:
    print(j,'...')     
```
